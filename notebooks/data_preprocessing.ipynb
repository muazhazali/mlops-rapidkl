```python
import pandas as pd
import glob
import os

# read csv from 01_raw and merge them into one
raw_data_dir = r"C:\GitHub\mlops-rapidkl\data\01_raw"
csv_files = glob.glob(os.path.join(raw_data_dir, "rapidrail_*_daily.csv"))

dfs = []
for file in csv_files:
    df = pd.read_csv(file)
    dfs.append(df)

merged_df = pd.concat(dfs, ignore_index=True)
print(f"Merged data shape: {merged_df.shape}")
merged_df.head()
```

```python
# count of data for each pair
pair_counts = merged_df.groupby(['origin', 'destination']).size().reset_index(name='count')
pair_counts.sort_values(by='count', ascending=False, inplace=True)
display(pair_counts.head(20))
```
